```python
import pandas as pd

df = pd.read_excel("topic_wise_pandas_related_sentences.xlsx")
df
```

```python
df = df[['Title','Topics', 'Correlation Score']]
df
```

```python
df = df.groupby('Topics')
df
```

```python
df.first()
```

```python
for i in range(1,27):
    mydf = df.get_group('Topic '+str(i))
    mydf = mydf.sort_values(by=['Correlation Score'], ascending=False)
    mydf.to_csv(f'Topic_{i}.csv', index=False)
```
